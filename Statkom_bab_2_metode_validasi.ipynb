#Uji Normalitas Jarque-Bera

Studi Kasus

Seorang dosen ingin mengetahui apakah nilai Ujian Statistika Komputasi dari 15 mahasiswa berdistribusi normal sebelum melakukan analisis inferensi statistik parametrik.

Data nilai mahasiswa:
[78, 82, 85, 80, 88, 90, 76, 84, 81, 79, 87, 83, 86, 77, 89]

```python
def jarque_bera_test(data):

    # Menghitung jumlah data
    n = len(data)

    # Menghitung rata-rata
    mean = sum(data) / n

    # Menghitung selisih data terhadap mean
    selisih = []
    for x in data:
        selisih.append(x - mean)

    # Menghitung momen ke-2
    m2 = 0
    for d in selisih:
        m2 += d**2
    m2 = m2 / n

    # Menghitung momen ke-3
    m3 = 0
    for d in selisih:
        m3 += d**3
    m3 = m3 / n

    # Menghitung momen ke-4
    m4 = 0
    for d in selisih:
        m4 += d**4
    m4 = m4 / n

    # Menghitung skewness
    skewness = m3 / (m2 ** 1.5)

    # Menghitung kurtosis
    kurtosis = m4 / (m2 ** 2)

    # Menghitung Jarque-Bera
    jb = (n / 6) * (skewness**2 + ((kurtosis - 3)**2) / 4)

    # Nilai kritis Chi-Square
    chi_square_critical = 5.991

    # Keputusan
    if jb > chi_square_critical:
        keputusan = "Tolak H0: data tidak berdistribusi normal"
    else:
        keputusan = "Gagal menolak H0: data berdistribusi normal"

    # Menampilkan hasil
    print("n =", n)
    print("mean =", mean)
    print("m2 =", m2)
    print("m3 =", m3)
    print("m4 =", m4)
    print("kurtosis =", kurtosis)
    print("skewness =", skewness)
    print("jb =", jb)
    print("keputusan =", keputusan)
```

```python
data = [78, 82, 85, 80, 88, 90, 76, 84, 81, 79, 87, 83, 86, 77, 89]

jarque_bera_test(data)
```

Interpretasi Hasil

Jika nilai JB < 5.991, maka H₀ gagal ditolak sehingga data nilai mahasiswa dapat dianggap berdistribusi normal.

Jika nilai JB > 5.991, maka H₀ ditolak sehingga data tidak berdistribusi normal.

#Uji Normalitas Kolmogorov-Smirnov (K-S)

Pengertian Uji Kolmogorov-Smirnov

Uji Kolmogorov-Smirnov (K-S) merupakan metode statistik yang digunakan untuk menguji apakah suatu sampel berasal dari populasi yang mengikuti distribusi tertentu, salah satunya distribusi normal. Uji ini dilakukan dengan membandingkan distribusi kumulatif empiris dari data sampel dengan distribusi kumulatif teoritis yang diharapkan.

Prinsip dasar Uji Kolmogorov-Smirnov adalah mencari selisih maksimum antara fungsi distribusi kumulatif empiris (Empirical Distribution Function/EDF) dan fungsi distribusi kumulatif teoritis (CDF). Semakin kecil selisih tersebut, semakin dekat distribusi data dengan distribusi teoritis yang diuji.

Hipotesis

H₀ : Data berdistribusi normal

H₁ : Data tidak berdistribusi normal

Rumus-Rumus Uji Kolmogorov-Smirnov

1. Menghitung Mean
\begin{equation*}
\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}
\end{equation*}
Keterangan:

$\bar{x}$ = rata-rata sampel \
$x_i$ = data ke-$i$ \
$n$ = jumlah observasi

2. Menghitung Standar Deviasi

$$s = \sqrt{\frac{\sum (x_i - \bar{x})^2}{n}}$$

Keterangan:

$s$ = standar deviasi

3. Menghitung Nilai Z


$$Z_i = \frac{x_i - \bar{x}}{s}$$

Keterangan:

\$Z_i$ = skor baku data ke-i


4. Fungsi Distribusi Empiris


$$F_o(x_i) = \frac{i}{n}$$

Keterangan:

$i$ = urutan data setelah diurutkan

5. Statistik Kolmogorov-Smirnov

$$D = \max |F_o(x_i) - F_t(x_i)|$$

Keterangan:

$F_o(x_i)$ = distribusi empiris

$F_t(x_i)$ = distribusi normal teoritis

$D$ = statistik Kolmogorov-Smirnov

Kriteria Keputusan

Pada taraf signifikansi 5\%:

$$D_{tabel} = \frac{1.36}{\sqrt{n}}$$

Jika

$$D > D_{tabel}$$

maka $H_0$ ditolak.

Jika

$$D \le D_{tabel}$$

maka $H_0$ gagal ditolak.

#Uji Homogenitas Varians (F Varians)

Studi Kasus

Seorang dosen ingin mengetahui apakah variasi nilai ujian mahasiswa dari dua kelas berbeda memiliki varians yang sama (homogen) sebelum dilakukan Uji T Dua Sampel Independen.

Data kelas A:

data1 = [78, 82, 85, 80, 88, 90]

Data kelas B:

data2 = [75, 95, 70, 85, 65, 90]

```python
def uji_F_varians(data1, data2, alpha=0.05):

    # Menghitung jumlah observasi
    n1 = len(data1)
    n2 = len(data2)

    # Menghitung rata-rata
    mean1 = sum(data1) / n1
    mean2 = sum(data2) / n2

    # Menghitung varians sampel pertama
    jumlah_kuadrat1 = 0

    for x in data1:
        jumlah_kuadrat1 += (x - mean1) ** 2

    varians1 = jumlah_kuadrat1 / (n1 - 1)

    # Menghitung varians sampel kedua
    jumlah_kuadrat2 = 0

    for x in data2:
        jumlah_kuadrat2 += (x - mean2) ** 2

    varians2 = jumlah_kuadrat2 / (n2 - 1)

    # Menghitung F hitung
    if varians1 > varians2:
        f_hitung = varians1 / varians2
        df1 = n1 - 1
        df2 = n2 - 1
    else:
        f_hitung = varians2 / varians1
        df1 = n2 - 1
        df2 = n1 - 1

    # Menghitung F tabel
    f_tabel = f.ppf(1 - alpha, df1, df2)

    # Keputusan
    if f_hitung > f_tabel:
        keputusan = "Tolak H0: varians tidak homogen"
    else:
        keputusan = "Gagal menolak H0: varians homogen"

    # Menampilkan hasil
    print("n1 =", n1)
    print("n2 =", n2)
    print("mean1 =", mean1)
    print("mean2 =", mean2)
    print("varians1 =", varians1)
    print("varians2 =", varians2)
    print("f_hitung =", f_hitung)
    print("f_tabel =", f_tabel)
    print("keputusan =", keputusan)
```

```python
data1 = [78, 82, 85, 80, 88, 90]
data2 = [75, 95, 70, 85, 65, 90]

uji_F_varians(data1, data2)
```

Interpretasi

Hipotesis:

$$H_0 : \sigma_1^2 = \sigma_2^2$$
$$H_1 : \sigma_1^2 \neq \sigma_2^2$$

Kriteria keputusan:

$$H_0 \text{ ditolak jika } F_{hitung} > F_{tabel}$$
$$H_0 \text{ gagal ditolak jika } F_{hitung} \le F_{tabel}$$

Jika $F_{hitung}$ lebih kecil dari $F_{tabel}$, maka kedua kelompok memiliki varians yang homogen sehingga asumsi homogenitas varians terpenuhi.

Uji Homogenitas Varians Levene

Pengertian Uji Levene

Uji Levene adalah metode statistik yang digunakan untuk menguji apakah dua atau lebih kelompok data memiliki varians yang sama (homogen). Uji ini dikembangkan oleh Howard Levene sebagai alternatif Uji F Varians yang lebih tahan terhadap pelanggaran asumsi normalitas dan keberadaan outlier.

Berbeda dengan Uji F Varians yang langsung membandingkan varians antar kelompok, Uji Levene mengubah data terlebih dahulu menjadi deviasi absolut terhadap ukuran pemusatan (umumnya mean atau median). Setelah itu dilakukan analisis varians (ANOVA) terhadap nilai deviasi tersebut.

Uji Levene banyak digunakan sebagai pengujian asumsi homogenitas varians sebelum melakukan Uji T, ANOVA, maupun metode statistik parametrik lainnya.

Hipotesis

$$H_0 : \sigma_1^2 = \sigma_2^2 = \cdots = \sigma_k^2$$

(Varians seluruh kelompok homogen)

$$H_1 : \text{Minimal terdapat satu varians kelompok yang berbeda}$$

(Varians tidak homogen)

Rumus-Rumus Uji Levene

1. Menghitung Rata-Rata Kelompok

$$\bar{X}_i = \frac{\sum_{j=1}^{n_i} X_{ij}}{n_i}$$

Keterangan:

$\bar{X}_i$ = rata-rata kelompok ke-i

$X_{ij}$ = data ke-j pada kelompok ke-i

$n_i$ = jumlah observasi kelompok ke-i

2. Menghitung Deviasi Absolut

$$Z_{ij} = |X_{ij} - \bar{X}_i|$$

Keterangan:

$Z_{ij}$ = deviasi absolut
    
$X_{ij}$ = data asli
   
$\bar{X}_i$ = rata-rata kelompok ke-i

3. Menghitung Rata-Rata Deviasi Tiap Kelompok

$$\bar{Z}_i = \frac{\sum_{j=1}^{n_i} Z_{ij}}{n_i}$$

4. Menghitung Rata-Rata Deviasi Keseluruhan

$$\bar{Z} = \frac{\sum_{i=1}^{k} \sum_{j=1}^{n_i} Z_{ij}}{N}$$

Keterangan:

$N$ = jumlah seluruh observasi
    
$k$ = jumlah kelompok

5. Statistik Uji Levene

$$W = \frac{(N - k) \sum_{i=1}^{k} n_i(\bar{Z}_i - \bar{Z})^2}{(k - 1) \sum_{i=1}^{k} \sum_{j=1}^{n_i} (Z_{ij} - \bar{Z}_i)^2}$$

Keterangan:

$W$ = statistik Levene

$N$ = jumlah seluruh data

$k$ = jumlah kelompok

Kriteria Pengujian

$$H_0 \text{ ditolak jika } W > F_{\alpha}(k - 1, N - k)$$
$$H_0 \text{ gagal ditolak jika } W \le F_{\alpha}(k - 1, N - k)$$

Uji Independensi Fisher's Exact Test

Pengertian Fisher's Exact Test

Fisher's Exact Test adalah metode statistik yang digunakan untuk menguji hubungan (independensi) antara dua variabel kategorik yang disajikan dalam tabel kontingensi berukuran 2×2.

Uji ini dikembangkan oleh Ronald A. Fisher dan digunakan sebagai alternatif Uji Chi-Square ketika ukuran sampel kecil atau ketika terdapat frekuensi harapan kurang dari 5 pada salah satu sel tabel kontingensi.

Berbeda dengan Uji Chi-Square yang menggunakan pendekatan distribusi asimtotik, Fisher's Exact Test menghitung peluang (probabilitas) secara tepat (exact probability) berdasarkan distribusi hipergeometrik sehingga hasilnya lebih akurat pada sampel kecil.

Hipotesis

$$H_0 : \text{Tidak terdapat hubungan antara kedua variabel}$$
$$H_1 : \text{Terdapat hubungan antara kedua variabel}$$

Rumus-Rumus Fisher's Exact Test

Misalkan tabel kontingensi 2×2 berikut:

image.png

1. Total Keseluruhan

$$n = a + b + c + d$$

2. Faktorial

$$n! = n \times (n - 1) \times (n - 2) \times \cdots \times 1$$

dengan

$$0! = 1$$

3. Probabilitas Fisher Exact Test

$$P = \frac{(a + b)!(c + d)!(a + c)!(b + d)!}{a!b!c!d!n!}$$

Keterangan:

$a, b, c, d$ = frekuensi observasi pada tabel $2 \times 2$

$n$ = jumlah seluruh observasi

$P$ = probabilitas Fisher Exact Test

Kriteria Pengujian

Jika

$$p-value< \alpha$$

maka:

$$H_0 \text{ ditolak}$$

Jika

$$p-value \ge \alpha$$

maka:

$$H_0 \text{ gagal ditolak}$$

```python
data = [
    [30, 20],
    [15, 35]
]

a = data[0][0]
b = data[0][1]
c = data[1][0]
d = data[1][1]

baris1 = a + b
baris2 = c + d

kolom1 = a + c
kolom2 = b + d

n = a + b + c + d

def faktorial(x):

    hasil = 1

    for i in range(1, x + 1):
        hasil *= i

    return hasil

p_value = (
    faktorial(baris1) *
    faktorial(baris2) *
    faktorial(kolom1) *
    faktorial(kolom2)
) / (
    faktorial(a) *
    faktorial(b) *
    faktorial(c) *
    faktorial(d) *
    faktorial(n)
)

alpha = 0.05

if p_value < alpha:
    keputusan = "Tolak H0: terdapat hubungan"
else:
    keputusan = "Gagal menolak H0: tidak terdapat hubungan"

print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)

print("baris1 =", baris1)
print("baris2 =", baris2)

print("kolom1 =", kolom1)
print("kolom2 =", kolom2)

print("n =", n)

print("p_value =", p_value)

print("keputusan =", keputusan)
```

Kesimpulan

Fisher's Exact Test digunakan untuk menguji hubungan antara dua variabel kategorik pada tabel kontingensi 2×2, terutama ketika ukuran sampel kecil atau terdapat frekuensi harapan yang rendah.

Dengan data:

data = [
    [30, 20],
    [15, 35]
]

nilai probabilitas dihitung menggunakan distribusi hipergeometrik:

$$P = \frac{(a + b)!(c + d)!(a + c)!(b + d)!}{a!b!c!d!n!}$$

Keputusan diambil berdasarkan nilai \textbf{\textit{p-value}}:

$$H_0 \text{ ditolak jika } p-value < 0.05$$
$$H_0 \text{ gagal ditolak jika } p-value\ge 0.05$$

Apabila p-value lebih kecil dari 0,05 maka terdapat hubungan antara kedua variabel yang diuji. Sebaliknya, jika p-value lebih besar atau sama dengan 0,05 maka tidak terdapat hubungan yang signifikan antara kedua variabel.
